# Data Structures

# List ==> []
* Collection of **ORDERED & CHANGEABLE** items.
* Elements can be of same / different types
* Duplicate members are allowed
* Elements are included within []

```python
# Creating list

# Method_1
li = ["1",2,"string"]
print(li)

# Method_2
li = list("124")
print(li)

# Method_3
li = list(i for i in range(10))
print(li)
```

# Tuple ==> ()
* Collection of **ORDERED & UNCHANGEABLE** items.
* Elements can be of same / different types
* Duplicate members are allowed
* Elements are included within ()

```python
# Creating tuple

# Method_1
t = ("1",2,"tuple")
print(t)

# Method_2
t = tuple("124")
print(t)

# Method_3
t = tuple(i for i in range(10))
print(t)
```

# Set ==> {}
* Collection of **UNORDERED & CHANGEABLE** items. **Also ==> UNINDEXED**
* Elements can be of same / different types
* Duplicate members are **NOT** allowed
* Elements are included within {}

```python
# Creating set

# Method_1
s = {"1",2,"set"}
print(s)

# Method_2
s = set("124")
print(s)

# Method_3
s = set(i for i in range(10))
print(s)
```

# Dictionary ==> {}
* Collection of **UNORDERED & CHANGEABLE** items. **Also ==> INDEXED**
* Elements can be of same / different types
* Duplicate members are **NOT allowed**
* Elements are included within {}

```python
# Creating Dictionary
# Format ==> {key1:value1, key2:value2,...}

d = {"brand":"Ford", 2:"Mustang", "year":1964}
print(d)
```
